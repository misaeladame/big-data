```python
import matplotlib.pyplot as plt
```

```python
#para ver plots juntos en la misma linea
plt.subplot2grid((3,3),(0,0))

data.survived.value_counts().plot(kind='bar')

plt.title('Los que sobrevivieron')
#manera mÃ¡s amable %
plt.subplot2grid((3,3),(0,1))
data.survived.value_counts(normalize=True).plot(kind='bar')
plt.title('Los que sobrevivieron en %')
plt.subplot2grid((3,3),(1,0))

data.survived.value_counts().plot(kind='bar')
plt.subplot2grid((3,3),(0,1))

data.survived.value_counts().plot(kind='bar')
plt.subplot2grid((3,3),(2,0))

data.survived.value_counts().plot(kind='bar')

#plt.show
```

```python
figura = plt.figure(figsize=(30,10)) #crea el canvas 30x10 pix
#para ver plots juntos en la misma linea

data.sex[data.survived==1].value_counts(normalize=True).plot(kind='bar')

plt.title('Los que sobrevivieron mujeres y hombres')
#plt.show()
```

```python
figura = plt.figure(figsize=(10, 5)) #crea el canvas 10x5 pix
#para ver plots juntos en la misma linea
data.pclass[data.survived == 1].value_counts(normalize = True).plot(kind='bar')
plt.title('Los que sobrevivieron por clase')
#plt.show()
```

```python
#relacion de la case con la edad
figura = plt.figure(figsize=(20,10))
for tipoBoleto in [1, 2, 3]:
    data.age[data.pclass == tipoBoleto].plot(kind='kde')
    
plt.legend(('1ra clase', '2da clase', '3ra clase'))
```
